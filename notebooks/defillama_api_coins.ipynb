## Get daily closes ('23:59:59')

DeFiLlama currently doesn't offer an API for bulk downloading historical prices. So I wrote a method to call its snapshot price API iteratively to get historical daily open (00:00:00) or close (23:59:59) prices. It is slow so use judiciously.

```python
df = obj.get_tokens_hist_prices(dd, start='2022-08-01', end='2022-09-01', type='close')
df.head()
```

```python
# calculate daily returns
simple_rets = df.pct_change()
simple_rets.head()
```

```python
# calculate the log returns
log_rets = np.log(simple_rets + 1)
log_rets.head()
```

```python
# calculate cumulative returns using the simple returns
cumu_rets_way1 = (1 + simple_rets).cumprod() - 1
cumu_rets_way1.head()
```

```python
# calculate cumulative returns using the log returns
cumu_rets_way2 = np.exp(log_rets.cumsum()) - 1
cumu_rets_way2.head()
```

```python
# check both methods give the same results
(cumu_rets_way1 - cumu_rets_way2).sum()
```
