## Get daily close prices ('23:59:59')

pretty fast because of the batch historical price data API

```python
df = obj.get_daily_open_close(dd, start='2022-11-01', end='2022-11-03', kind='close')
df.head()
```

```python
df = obj.get_daily_open_close(dd, start='2022-11-01', end='2022-12-26', kind='close')
print(df.head())
print(df.tail())
```

```python
type(df.index[0])
```

```python
# calculate daily returns
simple_rets = df.pct_change()
simple_rets.head()
```

```python
# calculate the log returns
log_rets = np.log(simple_rets + 1)
log_rets.head()
```

```python
# calculate cumulative returns using the simple returns
cumu_rets_way1 = (1 + simple_rets).cumprod() - 1
cumu_rets_way1.head()
```

```python
# calculate cumulative returns using the log returns
cumu_rets_way2 = np.exp(log_rets.cumsum()) - 1
cumu_rets_way2.head()
```

```python
# check both methods give the same results
(cumu_rets_way1 - cumu_rets_way2).sum()
```

## Get historical prices for different candle periods

for example: 15m, 1h, 4h, 1d, 1w

```python
# get 10 price points on the 15-minute chart for weth on arbitrum before '2022-12-25 20:15:00' 
df = obj.get_prices_at_regular_intervals(
    {'0x82aF49447D8a07e3bd95BD0d56f35241523fBab1':'arbitrum'},
    end = '2022-12-25 20:15:00', 
    end_format='%Y-%m-%d %H:%M:%S', 
    span = 10,
    period = '15m'
)
df.head()
```

```python

# get 5 price points before '2022-12-25 20:15:34' that are 1 hour apart
df = obj.get_prices_at_regular_intervals(
    dd,
    end = '2022-12-25 20:15:34', 
    end_format='%Y-%m-%d %H:%M:%S', 
    span = 5,
    period = '1h'
)
df.head()
```

```python
# get price points for the 5 weeks before '2022-12-25'
df = obj.get_prices_at_regular_intervals(
    dd,
    end = '2022-12-25', 
    end_format='%Y-%m-%d', 
    span = 5,
    period = '1w'
)
df.head()
```
